```python
# Data gathering
import pandas_datareader as pdr # type: ignore
import matplotlib.pyplot as plt # type: ignore


starting_date = '1990-01-01'
ending_date = '2024-08-05'

vix = pdr.DataReader('VIXCLS', 'fred', starting_date, ending_date)
vix = vix.dropna()
vix = vix.diff(periods=1, axis=0)
vix = vix.iloc[1: , :]

mean = vix["VIXCLS"].mean()

plt.plot(vix[-5500:], color='blue', linewidth=1.5, label="Change in VIX")
plt.axhline(y = mean, color = 'red', linestyle= 'dashed')
plt.grid()
plt.legend()
plt.show()
```
